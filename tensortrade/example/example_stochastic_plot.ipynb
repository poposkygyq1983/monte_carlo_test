```python
import random
import tensortrade.stochastic as sp
%matplotlib inline
```

### Geometric Brownian Motion


```python
data = sp.gbm(
    base_price=100,
    base_volume=5,
    start_date="2010-01-01",
    times_to_generate=1000,
    time_frame='1H'
)

data.close.plot()
```
